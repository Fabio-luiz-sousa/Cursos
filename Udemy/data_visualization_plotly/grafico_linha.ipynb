### Ploty Expres

```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
eixo_y=np.random.randint(low=100, high=5000, size=50)
```

```python
eixo_x=pd.date_range(start='2022-1-1', freq='D',periods=50)
```

```python
plt.figure(figsize=(8,5))
plt.plot(eixo_x,eixo_y)
plt.title('Minha figura')
plt.show()
```

```python
fig=px.line(x=eixo_x,y=eixo_y, title='Minha Figura')
fig.show()
```

### Alterando caractersticas do grfico

```python
N=50
x=np.linspace(0,1,N)
random_y=np.random.normal(30,2.5,N)
```

```python
linha=go.Scatter(x=x,y=random_y, mode='lines',name='Linha 1', line=dict(color='black'))
layout=go.Layout(title='Minha figura',xaxis=dict(title='Eixo X'),yaxis=dict(title='Eixo Y'))
fig=go.Figure(data=linha,layout=layout)
fig.show()
```

### Adicionando multiplas linhas

```python
N=50
x=np.linspace(0,1,N)
random_y1=np.random.normal(loc=30,scale=2.5,size=N)
random_y2=np.random.randint(low=0,high=50,size=N)
random_y3=np.random.uniform(low=0,high=20, size=N)
```

```python
linha1=go.Scatter(x=x,y=random_y1,mode='markers',name='Linha1',line=dict(color='red',width=2))
linha2=go.Scatter(x=x,y=random_y2,mode='lines+markers',name='Linha2',line=dict(color='black',width=1))
linha3=go.Scatter(x=x,y=random_y3,mode='lines',name='Linha3',line=dict(color='gray'))
data=[linha1,linha2,linha3]
layout=go.Layout(title='Minha Figura',xaxis=dict(title='Eixo X'),yaxis=dict(title='Eixo Y'))
fig=go.Figure(data,layout)
fig.show()
```

### Usando add_trace()

```python
N=50
x=np.linspace(0,1,N)
random_y1=np.random.normal(loc=30,scale=2.5,size=N)
random_y2=np.random.randint(low=0,high=50,size=N)
random_y3=np.random.uniform(low=0,high=20, size=N)
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=x,y=random_y1,mode='lines',line=dict(color='gray'),name='Linha1'))
fig.add_trace(go.Scatter(x=x,y=random_y2,mode='lines+markers',line=dict(color='darkslategray',width=2),name='Linha2',marker=dict(size=5,color='green')))
fig.add_trace(go.Scatter(x=x,y=random_y3,mode='lines',line=dict(color='red',width=2,dash='dot'),name='Linha3'))

fig.update_layout(plot_bgcolor='white',title='Grfico de Linhas',font_family='Arial',font_color='black',title_font_color='gray',
xaxis=dict(title='Eixo X'),yaxis=dict(title='Eixo Y'),legend_title='Linhas',legend_title_font_color='dimgray')

fig.show()
```
