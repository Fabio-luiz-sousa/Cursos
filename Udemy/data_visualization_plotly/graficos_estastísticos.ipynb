### Histogramas

```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
data={'A':np.random.normal(10,0.5,10_000),'B':np.random.normal(10,0.5,10_000)}
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
fig=px.histogram(df,x='A',title='Histograma variavel A',nbins=20)
fig.update_layout(xaxis=dict(title='Variavel A'),yaxis=dict(title='Frequência'))
fig.show()
```

```python
fig=px.histogram(df,x='B',title='Histograma variavel B',nbins=20)
fig.update_layout(xaxis=dict(title='Variavel B'),yaxis=dict(title='Frequência'))
fig.show()
```
